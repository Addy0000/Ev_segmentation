# The model

we are using k means clustering

**Importing machine learning libraries**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

**The k_means model for demographic segmentation:-**

```python
X_scaled = StandardScaler().fit_transform(customer_data[['Age', 'Salary', 'Total Salary','Price']])
X_scaled = pd.DataFrame(X_scaled,columns=['Age', 'Salary', 'Total Salary','Price'])
x = X_scaled.to_numpy()
X_scaled
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++',
                    max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('k values')
plt.ylabel('WCSS') #within cluster sum of squares
plt.tight_layout()
plt.show()
```

we use the k value where the slop gets more slant

**We will keep k as 4, hence K = 4**

```python
kmeans = KMeans(n_clusters = 4)
label = kmeans.fit_predict(X_scaled)
print(label)
```

```python
labels = pd.DataFrame(label,columns=['K value'])
labels.head()
```

```python
final = customer_data[['Age', 'Salary', 'Total Salary','Price']].copy()
final = pd.concat([final,labels],axis=1) #the final dataset including all the relevent information.
final.head()
```

```python
print(kmeans.cluster_centers_) #These are the centroids for all the clusters.
```

```python
sns.pairplot(final,x_vars = ['Age','Salary','Total Salary','Price'], y_vars = ['Age','Salary','Total Salary','Price'], hue='K value')
```

**So, in conclusion these are all the possible clusters or segments that we have using demographic segmentation.**
